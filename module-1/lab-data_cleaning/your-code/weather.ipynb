## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
df = pd.read_csv('../weather-raw.csv')
display(df)
```

```python
print('Nº Null:\n',df.isnull().sum())
```

```python
'''
Elimino las columnas que no tienen datos por lo que las considero columnas erroneas
'''

df= df[['id', 'year', 'month', 'element', 'd1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd10', 'd11', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd23', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31']]
display(df)
```

```python
df = df.fillna(0)
#display(df)
variable = df[['id','year','month', 'element']]
```

```python
variable
```

```python
Temperaturas  =df.drop(columns=['id','year','month', 'element'])
```

```python
Temperaturas
```

```python
temperatures_fina = []
for i in range(22):
    media = []
    for e in Temperaturas.iloc[i]:
        if e != 0:
            media.append(e)   
    temperatures_fina.append(sum(media)/len(media))
```

```python
df_limpio = variable
df_limpio['Temp_Media_dias'] = temperatures_fina
display(df_limpio)


```
